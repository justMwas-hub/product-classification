```python
data_1 = df[["Product Title", "Category Label"]]
```

```python
data_1
```

```python
X = data_1["Product Title"]
y = data_1["Category Label"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
help(TfidfVectorizer())
```

```python
tfidf = TfidfVectorizer(stop_words="english")
```

```python
tfidf.fit(X_train)
```

```python
X_train_tfidf = tfidf.transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
X_train_tfidf
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(X_train_tfidf,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(max_iter=1000)
log.fit(X_train_tfidf,y_train)
```

```python
from sklearn.svm import LinearSVC
l_svc = LinearSVC()
l_svc.fit(X_train_tfidf,y_train)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train_tfidf,y_train)
```

```python
from sklearn.metrics import plot_confusion_matrix, classification_report
```

```python
def report(model):
    preds = model.predict(X_test_tfidf)
    print(classification_report(y_test,preds))
    plot_confusion_matrix(model,X_test_tfidf,y_test,xticks_rotation='vertical')
```

```python
print("NB MODEL")
report(nb)
```

```python
print("LOG MODEL")
report(log)
```

```python
print("L_SVC MODEL")
report(l_svc)
```

```python
print("SVC MODEL")
report(svc)
```

```python
from sklearn.naive_bayes import ComplementNB
comp_nb = ComplementNB()
comp_nb.fit(X_train_tfidf,y_train)
```

```python
print("COMP NB")
report(comp_nb)
```
